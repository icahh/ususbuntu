```python
import pandas as pd  # Import library pandas lalu inisialisasikan sebagai pd
import numpy as np  # Import library numpy kemudian inisialisasikan sebagai np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

HeartDisease = pd.read_csv('DataDummy.csv')  # Input data
HeartDisease  # Menampilkan data
```

```python
def prob_prior(HeartDisease, target):
    label_kelas = sorted(list(HeartDisease[target].unique()))
    prior=[]
    for i in label_kelas:
        prior.append(len(HeartDisease[HeartDisease[target]==i])/len(HeartDisease))
    return prior
```

```python
def likelihood_categorial(HeartDisease, nam_var, nilai_var, target, label):
    HeartDisease=HeartDisease[HeartDisease[target]==label]
    p_x_dgsyrt_y = len(HeartDisease[HeartDisease[nam_var]==nilai_var])/len(HeartDisease)
    return p_x_dgsyrt_y
```

```python
def categorical (HeartDisease, age, target):
    nama_kolom = list(HeartDisease.columns)[:-1]
    prior = prob_prior(HeartDisease, target)
    y_predd = []
    for Target in target:
        labels = sorted(list(HeartDisease[target].unique()))
        likelihood = [1]*len(labels)
        for j in range (len(labels)):
            for i in range (len(nama_kolom)):
                likelihood[j] *= likelihood_categorical(HeartDisease, nama_kolom[i], Target[i], age, labels[j])
        post_prob = [1]*len(labels)
        for j in range(len(labels)):
            post_prob[j]=likelihood[j]*prior[j]
        y_predd.append(np.argmax(post_prob))
    return np.array(y_predd)
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import BayesianRidge
   
# Loading dataset
dataset = load_boston()
X, y = dataset.data, dataset.target
   
# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
   
# Creating and training model
model = BayesianRidge()
model.fit(X_train, y_train)
   
# Model making a prediction on test data
prediction = model.predict(X_test)
   
# Evaluation of r2 score of the model against the test set
print(f"r2 Score Of Test Set : {r2_score(y_test, prediction)}")
```
